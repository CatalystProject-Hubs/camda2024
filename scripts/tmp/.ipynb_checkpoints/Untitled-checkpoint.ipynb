# Timeline

```python
data = pd.read_csv('/Users/joaquintripp/Documents/GitHub/camda24/00_origin_data/limpio_complete.csv')
cat = pd.read_csv('/Users/joaquintripp/Documents/GitHub/camda24/00_origin_data/gen2_CIE10.csv')
data['total_data'] = data.apply(lambda x: len([v for v in x if v >0]),axis = 1)
data.head()
```

```python
cat
```

```python
def plotTimeline(ID:int, save:bool=False) -> object:
    if ID not in list(data['id']):
        return 'ID not found'
    # ============================================================= preprocess
    sexo = data.loc[data['id']==ID,'sexo'].values[0]
    sexo = 'Male' if sexo == 1 else sexo
    sexo = 'Female' if sexo == 2 else sexo
    sexo = 'Unknow' if sexo == 3 else sexo

    dxs = [col for col in data.columns if col not in ['sexo', 'id', 'H35_bin', 'I20-I25_bin', 'Z89_bin', 'N18_bin','total_data']]
    aux = data.loc[data['id']==ID, dxs].T.reset_index()
    aux.columns = ['cie','age']
    aux = aux.loc[aux['age']>0,:].sort_values(by='age')
    aux = pd.merge(
        aux,
        cat[['BPS_PATHOLOGY','CIE-10']],
        left_on = 'cie',
        right_on = 'CIE-10',
        how = 'left'
    )
    aux['BPS_PATHOLOGY'] = aux['CIE-10']+' - '+aux['BPS_PATHOLOGY']

    # ============================================================= Insights
    insights = aux.loc[aux['cie'].isin(['H35','I20-I25','Z89','N18','E10-E14'])]
    insights_dict = {}

    for index,row in insights.iterrows():
        in_cie = row['cie']
        age = row['age']
        bps = row['BPS_PATHOLOGY']

        insights_dict[in_cie] = [age,bps]


    plot_data = aux
    # Create the plot
    plt.figure(figsize=(12, 6))
    plt.plot(plot_data["age"],range(len(plot_data["age"])),'o-', label='BPS_PATHOLOGY', c='#5A72A0', zorder=2)
    plt.yticks(range(len(plot_data["BPS_PATHOLOGY"])), plot_data["BPS_PATHOLOGY"])
    plt.xticks(np.arange(min(plot_data['age']), max(plot_data['age']) + 1, 1))
    plt.xlabel('Age')
    plt.ylabel('')
    plt.title(f'Patient {ID} - {sexo}', fontsize = 16)
    plt.grid(True)

    # highlight insights
    if len(insights_dict) > 0:
        for k,val in insights_dict.items():
            y_index = plot_data.loc[plot_data["cie"]==k].index[0]
            plt.scatter(val[0],y_index, s=150, c='#1A2130', zorder=3, alpha=1)
            plt.annotate(k, xy=(val[0], y_index), xytext=(val[0] - 4, y_index + 0.2),
                         arrowprops=dict(facecolor='black', arrowstyle='-'))

    plt.tight_layout()
    if save: 
        plt.savefig(f'{ID}_timeline.png', dpi = 400)
    return plt.show()
```
